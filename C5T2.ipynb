```python
# check if null values in dataframe
credit.isnull().values.any()
```

```python
# Drop ID feature
credit_ready = credit.drop(['ID'], axis=1)
```

```python
# Rename "default payment next month" column
credit_ready.rename(columns={'default payment next month': 'default'}, inplace=True)
```

```python
# Convert sex, education, marriage, pay_0 trhough pay_5, and C to Categorical variables
credit_ready['SEX'] = credit_ready.SEX.astype('category')
credit_ready['EDUCATION'] = credit_ready.EDUCATION.astype('category')
credit_ready['MARRIAGE'] = credit_ready.MARRIAGE.astype('category')
credit_ready['PAY_0'] = credit_ready.PAY_0.astype('category')
credit_ready['PAY_2'] = credit_ready.PAY_2.astype('category')
credit_ready['PAY_3'] = credit_ready.PAY_3.astype('category')
credit_ready['PAY_4'] = credit_ready.PAY_4.astype('category')
credit_ready['PAY_5'] = credit_ready.PAY_5.astype('category')
credit_ready['PAY_6'] = credit_ready.PAY_6.astype('category')
credit_ready['default'] = credit_ready.default.astype('category')
```

```python
credit_ready.info()
```

```python
# Visualizations
import matplotlib.pyplot as plt
header = credit_ready.dtypes.index
print(header)
```

```python
# credit limit distribution
plt.hist(credit_ready['LIMIT_BAL'],6)
plt.show()
```

```python
# age distribution
plt.hist(credit_ready['AGE'],6,color = "skyblue", ec="skyblue")
plt.show()
```

```python
# credit limit line plot
plt.plot(credit_ready['LIMIT_BAL'], color='green')
plt.show()
```

```python
# scatter plot
x = credit_ready['PAY_0']
y = credit_ready['PAY_2']
plt.scatter(x,y)
plt.show()
```

```python
# Box plot
A = credit_ready['BILL_AMT1']
plt.boxplot(A,0,'gD')
plt.show()
```

```python
# Correlation
# The billing amount appear to correlated 
corrMat = credit_ready.corr()
print(corrMat)
```

```python
# Covariance
covMat = credit_ready.cov()
print(covMat)
```

```python
# Nunmber of creditholders by gender -  majority female
credit_ready.groupby('SEX')['SEX'].count()
```

```python
# Number of clients by marriage status - large majority with at least University level
credit_ready.groupby('MARRIAGE')['MARRIAGE'].count()
```

```python
# Repayment status for the most recent month - Over 6000 crediholders with payment delay of at least 1 month
import seaborn as sns
fg = sns.factorplot('PAY_0', data=credit_ready, kind='count', aspect=1.5)
fg.set_xlabels('PAY_0')
```

```python
# Default status by gender - sligthly larger percentage of male in default
g = sns.factorplot('SEX', data=credit_ready, hue='default', kind='count', aspect=1.75)
g.set_xlabels('SEX')
```

```python
# Default status by marriage 
h = sns.factorplot('MARRIAGE', data=credit_ready, hue='default', kind='count', aspect=1.75)
h.set_xlabels('MARRIAGE')
```

```python
# Age distribution with seaborn -  Majority within 20 and 40 years
age = credit_ready['AGE']
age_dist = sns.distplot(age)
age_dist.set_title("Distribution of Creditholders' Ages")
```

```python
# Distribution of age by gender -  both men and women have similar distribution
fig = sns.FacetGrid(credit_ready, hue='SEX', aspect=4)
fig.map(sns.kdeplot, 'AGE', shade=True)
oldest = credit_ready['AGE'].max()
fig.set(xlim=(0,oldest))
fig.set(title='Distribution of Age Grouped by Gender')
fig.add_legend()
```

```python
# number of married and single women in default is about equal. Same for men
table = pd.crosstab(index=[credit_ready.default], columns=[credit_ready.SEX,credit_ready.MARRIAGE])
table.columns.set_levels(['Female', 'Male'], level=0, inplace=True)
table.columns.set_levels(['Others','Married', 'Single','Divorced'], level=1, inplace=True)

table
```
